CALCULATION OF TOTAL MARKS AND ASSIGNMENT OF GRADES:
Criteria:

Grade A: 100-85

Grade B: 84-69

Grade C: 68-53

Grade D: 52-42

Grade E: 41-35

Grade F: 34-0

```python
df['mean']= df.mean(axis=1).round().astype('int64')
df
```

```python
grade=[]
for i in df['mean']:
    if i>=85 and i<=100:
        grade.append('A')
    elif i>=69 and i<=84:
        grade.append('B')
    elif i>=53 and i<=68:
        grade.append('C')
    elif i>=42 and i<=52:
        grade.append('D')
    elif i>=35 and i<=41:
        grade.append('E')
    else:
        grade.append('F')
df['grades'] = grade
df
```

```python
plt.title('Grades vs Ethinicity', fontsize=20)
sns.countplot( hue='grades', x= 'race/ethnicity', data=df)
plt.show()
```

```python
sns.set()
plt.title('Grades vs Parental Level Of Education', fontsize=20)
sns.countplot(x= 'parental level of education', hue='grades',  data=df)
plt.show()
```

```python
plt.title('Grade vs Gender', fontsize=20)
sns.countplot( x='gender', hue= 'grades', data=df)
plt.show()
```

```python
sns.countplot( x = 'lunch',  hue='grades', data=df)
plt.title('Grades vs Lunch', fontsize=20)
sns.set()
plt.show()
```

```python
sns.set()
plt.title('Grades vs Test Preparation Course', fontsize=20)
sns.countplot( hue='grades', x= 'test preparation course', data=df)
plt.show()
```
